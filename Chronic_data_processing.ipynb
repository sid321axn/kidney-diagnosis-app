```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('chronic.csv')
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df=df.replace('?', np.nan)
```

```python
df
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
df['Age'] = df['Age'].fillna(0).astype(int)
```

```python
df.dtypes
```

```python
df['Age']=df['Age'].replace(0, np.mean(df['Age']))
```

```python
df.describe()
```

```python
df['BP'] = df['BP'].fillna(0).astype(int)
```

```python
df['BP']=df['BP'].replace(0, np.mean(df['BP']))
```

```python
df
```

```python
df['RBC'].fillna(method='ffill',inplace=True)
```

```python
df['RBC'].fillna(method='bfill',inplace=True)
```

```python
df['PC'].fillna(method='ffill',inplace=True)
df['PCC'].fillna(method='ffill',inplace=True)
df['bacteria'].fillna(method='ffill',inplace=True)
df['specific_gravity'].fillna(method='ffill',inplace=True)
df['anemia'].fillna(method='ffill',inplace=True)
df['pedal_edema'].fillna(method='ffill',inplace=True)
df['apetite'].fillna(method='ffill',inplace=True)
df['CAD'].fillna(method='ffill',inplace=True)
df['diabetes'].fillna(method='ffill',inplace=True)
df['htn'].fillna(method='ffill',inplace=True)
df['Albumin'].fillna(method='ffill',inplace=True)
```

```python
df['sugar'].fillna(method='ffill',inplace=True)
```

```python
df.isna().sum()
```

```python
df['Blood_Gluc_rand'] = df['Blood_Gluc_rand'].fillna(0).astype(int)
df['Blood_Gluc_rand']=df['Blood_Gluc_rand'].replace(0, np.mean(df['Blood_Gluc_rand']))

df['Blood_Urea'] = df['Blood_Urea'].fillna(0).astype(float)
df['Blood_Urea']=df['Blood_Urea'].replace(0, np.mean(df['Blood_Urea']))

df['Serum_Cr'] = df['Serum_Cr'].fillna(0).astype(float)
df['Serum_Cr']=df['Serum_Cr'].replace(0, np.mean(df['Serum_Cr']))

df['sodium'] = df['sodium'].fillna(0).astype(float)
df['sodium']=df['sodium'].replace(0, np.mean(df['sodium']))

df['potassium'] = df['potassium'].fillna(0).astype(float)
df['potassium']=df['potassium'].replace(0, np.mean(df['potassium']))

df['hemoglobin'] = df['hemoglobin'].fillna(0).astype(float)
df['hemoglobin']=df['hemoglobin'].replace(0, np.mean(df['hemoglobin']))


df['packed_cell_volume'] = df['packed_cell_volume'].fillna(0).astype(float)
df['packed_cell_volume']=df['packed_cell_volume'].replace(0, np.mean(df['packed_cell_volume']))

df['wbc_cnt'] = df['wbc_cnt'].fillna(0).astype(int)
df['wbc_cnt']=df['wbc_cnt'].replace(0, np.mean(df['wbc_cnt']))

df['rbc_cnt'] = df['rbc_cnt'].fillna(0).astype(int)
df['rbc_cnt']=df['rbc_cnt'].replace(0, np.mean(df['rbc_cnt']))
```

```python
df['packed_cell_volume']=df['packed_cell_volume'].replace('?', np.nan)
```

```python
df['packed_cell_volume']
```

```python
df.to_csv('chronic_cleaned.csv')
```
